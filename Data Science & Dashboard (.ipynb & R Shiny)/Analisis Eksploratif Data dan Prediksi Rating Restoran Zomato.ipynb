### Identifikasi Outliers pada atribut 'cost' & 'votes'

```python
df.describe()
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    
    # Hitung IQR
    IQR = Q3 - Q1
    
    # Batas bawah dan batas atas untuk deteksi outlier
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Mendapatkan data yang merupakan outlier
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    return len(outliers)

variables = ['cost', 'votes']
for var in variables:
    num_outliers = detect_outliers_iqr(df, var)
    print(f'Jumlah outlier pada {var}: {num_outliers}')
```

### c. Apakah tema atau konsep restoran penting dalam menarik pelanggan?

### e. Apakah ada kemiripan antara wilayah-wilayah tertentu di Bengaluru berdasarkan jenis makanan yang disajikan?

```python
df['cuisines_list'] = df['cuisines'].str.split(',')
df_exploded = df.explode('cuisines_list')

cuisine_pivot = pd.pivot_table(df_exploded, 
                               index='location', 
                               columns='cuisines_list', 
                               aggfunc='size', 
                               fill_value=0)

cuisine_pivot.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
cuisine_pivot['Cluster'] = kmeans.fit_predict(cuisine_pivot)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(cuisine_pivot, method='ward')

plt.figure(figsize=(12, 8))
dendrogram(linked, 
           orientation='top', 
           labels=cuisine_pivot.index, 
           distance_sort='descending', 
           show_leaf_counts=True)

plt.title('Dendrogram of Cuisines by Location')
plt.xlabel('Locations')
plt.ylabel('Distance')
plt.show()
```

Berdasarkan dendrogram yang ditampilkan, dapat disimpulkan bahwa ada kemiripan antara wilayah-wilayah tertentu di Bengaluru dalam hal jenis makanan yang disajikan. Lokasi-lokasi yang tergabung dalam satu kelompok menunjukkan bahwa mereka menawarkan jenis makanan yang serupa, yang mungkin disebabkan oleh preferensi lokal, budaya kuliner, atau keberadaan restoran dengan konsep yang sama.

Garis hijau merepresentasikan kelompok lokasi yang memiliki kemiripan tertinggi dalam penawaran makanan. Artinya, restoran di kelompok ini kemungkinan besar menyajikan jenis makanan yang serupa atau memiliki karakteristik yang sejalan. Di sisi lain, garis oranye juga menunjukkan kelompok yang mirip, tetapi dengan tingkat kemiripan yang sedikit lebih rendah dibandingkan dengan kelompok hijau. Meskipun kelompok oranye memiliki kesamaan dalam jenis makanan, mereka menunjukkan perbedaan yang cukup signifikan dibandingkan dengan kelompok hijau.

Garis biru yang menghubungkan kelompok oranye dan hijau mengindikasikan adanya hubungan yang signifikan antara kedua kelompok ini, meskipun mereka tetap terklasifikasi terpisah. Ini menunjukkan bahwa restoran di kedua kelompok tersebut mungkin menawarkan beberapa jenis masakan yang sama atau memiliki elemen kuliner yang saling berkaitan.

### g. Jika dua wilayah serupa, apakah ini berarti wilayah tersebut dihuni oleh kelompok masyarakat yang mirip, atau tempat tersebut menarik jenis orang tertentu?

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Mengambil data yang relevan
data = df[['votes', 'rate', 'cuisines']]  # Contoh kolom
data = pd.get_dummies(data, columns=['cuisines'])  # Encoding Cuisines
data_scaled = StandardScaler().fit_transform(data)

# K-Means Clustering
kmeans = KMeans(n_clusters=3)
data['Cluster'] = kmeans.fit_predict(data_scaled)

# Visualisasi
plt.figure(figsize=(10, 8))
plt.scatter(data['votes'], data['rate'], c=data['Cluster'], cmap='viridis', alpha=0.5)
plt.title('Cluster Analysis of Regions Based on Restaurant Characteristics')
plt.xlabel('Votes')
plt.ylabel('Average Rating')
plt.colorbar(label='Cluster')
plt.show()
```

Dari analisis ini, dapat disimpulkan bahwa wilayah-wilayah yang serupa dalam hal jumlah suara dan rating rata-rata kemungkinan besar dihuni oleh kelompok masyarakat yang mirip. Kluster dengan jumlah suara tinggi menunjukkan bahwa wilayah tersebut menarik jenis pelanggan tertentu yang mencari pengalaman makan berkualitas tinggi. Sementara itu, wilayah yang terpisah dalam kluster dengan jumlah suara rendah dan rating lebih rendah mungkin menarik kelompok yang berbeda atau memiliki penawaran yang lebih terbatas.
